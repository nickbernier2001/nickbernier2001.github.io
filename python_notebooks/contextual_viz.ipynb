```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import altair as alt
import ipywidgets
```

```python
df = pd.read_csv('Census_Data_-_Languages_spoken_in_Chicago__2008___2012.csv')
df
```

```python
columns = df['Community Area Name'].values.tolist()
columns.remove('Chicago')
languages = df.columns.tolist()
languages.remove('Community Area')
languages.remove('Community Area Name')
languages.remove('PREDOMINANT NON-ENGLISH LANGUAGE (%)')

@ipywidgets.interact(CommunityName = columns)
def make_bar_plot(CommunityName):
    x = languages
    y = df[df['Community Area Name'] == CommunityName][languages].values[0]
    plt.xticks(rotation = 90)
    plt.tick_params(axis='x', labelsize=8)
    plt.title(CommunityName)
    plt.xlabel('Language')
    plt.ylabel('Number of Speakers')
    plt.bar(x, y)
    plt.show()
```
